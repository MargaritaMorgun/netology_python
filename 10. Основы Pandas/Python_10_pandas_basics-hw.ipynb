**Задание 2**

По данным файла power.csv посчитайте суммарное потребление стран Прибалтики (Латвия, Литва и Эстония) категорий 4, 12 и 21 
за период с 2005 по 2010 года. Не учитывайте в расчетах отрицательные значения quantity.

```python
power = pd.read_csv('power.csv')
power.head()
```

```python
power[ (
    ((power['country'] == 'Lithuania') | (power['country'] == 'Latvia') | (power['country'] == 'Estonia')) & 
    ((power['category'] == 4) | (power['category'] == 12) | (power['category'] == 21)) & 
    ((2005 <= power['year']) & (power['year']<= 2010)) &
    (power['quantity'] >= 0)
)]['quantity'].sum()
```
